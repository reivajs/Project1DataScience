## Problema 1

```python
# creamos una lista con los dias de ventas
dias = ["17-05-2022","18-05-2022","19-05-2022","20-05-2022","21-05-2022"]
# generamos los productos comprado para cada dia
productos = ["huevo","pollo","carne","jamon","leche"]
# generamos las unidades totales compradas
unidades = [12,6,5,3,20]
# generamos el precio de cada unidad
preciou = [22,69,80,12,24]

```

```python
# ahora creamos un dataframe donde el indice sea la fecha
dataframe = pd.DataFrame(list(zip(dias,productos,unidades,preciou))
                         , columns = ['Días','Producto',"Unidades","Precio Unitario"],index=dias)
```

```python
# creamos una nueva columna con el precio total
dataframe["Precio total"] = dataframe["Unidades"]*dataframe["Precio Unitario"]
```

```python
# vemos los resultados
dataframe
```

```python
# ahora creamos una nueva columna agregando el 5% de descuento 
dataframe["Precio descuento"] = dataframe["Precio total"]*0.95
```

```python
# vemos los resultados
dataframe
```

## problema 2

```python
# creamos el diccionario con los clientes mensuales de la empresa
# se usaras los ingresos mensuales de los clientes  para cierta empresa
clientes = {"cliente 1":17000 , "cliente 2":34000, "cliente 3": 10000, "cliente 4":18000, "cliente 5":7000,
           "cliente 6":22000,"cliente 7":9000,"cliente 8":45000,"cliente 9":6000,"cliente 10":25000}
```

```python
# calculamos el ingreso minimo, mximo y la media
maximo = max(clientes, key=clientes.get) 
minimo = min(clientes, key=clientes.get)
```

```python
maximo = [clientes[maximo]]
minimo = [clientes[minimo]]
```

```python
a=tuple(clientes.values())
b=len(a)
promedio=0
for val in a:
    promedio+=val
promedio = [promedio/b]
```

```python
# guardamos valores en dataframe
ingresos = pd.DataFrame(list(zip(maximo,minimo,promedio))
                         , columns = ['Ingreso máximo','Ingreso mínimo',"Ingreso promedio"])
```

## problema 3

```python
# guardamos los valores en una lista
mes = ["Enero","Febrero","Marzo","Abril","Mayo"]
ventas = [30500,35600,28300,33900,42500]
gasto = [22000,23450,18100,35700,32450]
```

```python
# creamos el dataframe 
problema3 = pd.DataFrame(list(zip(mes,ventas,gasto)), columns = ['Mes','Ventas',"Gastos"])
```

```python
# vemos el dataframe 
problema3
```

```python
# creamos las diferentes columnas
problema3["Columna 1"] = (problema3.Gastos/problema3.Ventas)*100
```

```python
# creaos la nueva columna vacia
problema3["Columna 2"] = " "
```

```python
# la rellenamos dados los daots
for i in range(0,len(problema3["Columna 2"])):
    if problema3["Ventas"][i] >= 30000:
        problema3["Columna 2"][i] = "Meta superada"
    else:
        problema3["Columna 2"][i] = "Meta no superada"
```

```python
# vemos los resultados
problema3
```

```python
# creaos la nueva columna vacia
problema3["Columna 3"] = " "
```

```python
# rellenamos la nueva columna
for i in range(0,len(problema3["Columna 3"])):
    if problema3["Gastos"][i] > problema3["Ventas"][i]:
        problema3["Columna 3"][i] = problema3["Gastos"][i] - problema3["Ventas"][i]
    else:
        problema3["Columna 3"][i] = 0
```

```python
# vemos los resultados
problema3
```

```python
# Total de venta en MArzo, Abril y mayo
# filtramos esos meses
nuevo = problema3[problema3.Mes.isin(["Marzo", "Abril","Mayo"])]
```

```python
# vemos resultados
nuevo
```

```python
# calcuamos las ventas 
ventas = sum(nuevo.Ventas)
```

```python
print("Ventas ultimos 3 meses:",ventas)
```

```python
# filtramos los gastos de febrero a abril
nuevo2 = problema3[problema3.Mes.isin(["Febrero", "Marzo","Abril"])]

gastos = sum(nuevo2.Gastos)
```

```python
print("Los gatos para los meses de interes son:",gastos)
```
